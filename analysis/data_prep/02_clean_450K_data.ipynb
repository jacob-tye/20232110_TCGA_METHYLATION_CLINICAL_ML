```python
import pandas as pd
```

```python
metadata = pd.read_csv('downloads/metadata.csv')
```

```python
len(metadata)
```

```python
metadata.head()
```

```python
metadata["file_path"] = metadata["file_id"] + "/" + metadata["file_name"]
```

```python
filename_to_id = dict(
    zip(metadata["file_path"], metadata["cases__samples__submitter_id"])
)
```

```python
all_data = {}
meth_ref = None

for filename, sample_id in filename_to_id.items():
    data = pd.read_csv(f"downloads/{filename}", sep="\t", header=None)
    all_data[sample_id] = data[1]
    if meth_ref is None:
        meth_ref = data[0]
```

```python
all_meth_df = pd.DataFrame(all_data)
all_meth_df["id"] = meth_ref
cols = all_meth_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
all_meth_df = all_meth_df[cols]
all_meth_df
```

```python
all_meth_df.to_feather("../../data/methylation/hm450Kmeth_data.feather")
```
