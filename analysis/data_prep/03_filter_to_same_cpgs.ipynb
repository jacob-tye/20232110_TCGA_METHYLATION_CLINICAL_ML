```python
import pandas as pd
```

```python
SAMPLES_METHYLATION_FILE = "/uufs/chpc.utah.edu/common/home/u0914269/clement/projects/20230828_tcga_methylation/side_projects/20232110_TCGA_METHYLATION_CLINICAL_ML/data/methylation/hm450Kmeth_data.feather"
SUPPLEMENTAL_NORMAL_METHYLATION_FILE = "/uufs/chpc.utah.edu/common/home/u0914269/clement/projects/20230828_tcga_methylation/side_projects/20232110_TCGA_METHYLATION_CLINICAL_ML/data/methylation/normals/normal_samples_450K.tsv"
ALL_SAMPLES_FILE = "/uufs/chpc.utah.edu/common/home/u0914269/clement/projects/20230828_tcga_methylation/side_projects/20232110_TCGA_METHYLATION_CLINICAL_ML/data/methylation/all_samples_450K.tsv"
ALL_CPGS_FILE = "/uufs/chpc.utah.edu/common/home/u0914269/clement/projects/20230828_tcga_methylation/side_projects/20232110_TCGA_METHYLATION_CLINICAL_ML/data/methylation/all_cpgs_450K.tsv"
```

```python
samples_methylation = pd.read_feather(
    SAMPLES_METHYLATION_FILE
)

supplemental_normal_methylation = pd.read_csv(
    SUPPLEMENTAL_NORMAL_METHYLATION_FILE,
    sep="\t"
)
```

```python
samples = [column for column in supplemental_normal_methylation.columns if column != 'key']
```

Modify the scale of the methylation to be between 0 and 1

```python
supplemental_normal_methylation[samples] = (
    supplemental_normal_methylation[samples] / 100
)
```

```python
supplemental_normal_methylation.describe()
```

```python
samples_methylation.describe()
```

```python
shared_cpgs = set(samples_methylation['id']).intersection(set(supplemental_normal_methylation['key']))
```

```python
len(shared_cpgs)
```

```python
filtered_samples_methylation = samples_methylation[samples_methylation['id'].isin(shared_cpgs)]
filtered_supplemental_normal_methylation = supplemental_normal_methylation[supplemental_normal_methylation['key'].isin(shared_cpgs)]
```

```python
filtered_samples_methylation.shape
```

```python
filtered_supplemental_normal_methylation.shape
```

```python
sorted_filtered_samples_methylation = filtered_samples_methylation.set_index('id').loc[filtered_supplemental_normal_methylation['key']].reset_index()
```

```python
sorted_filtered_samples_methylation
```

```python
sorted_filtered_samples_methylation.to_feather(SAMPLES_METHYLATION_FILE)
filtered_supplemental_normal_methylation.to_csv(SUPPLEMENTAL_NORMAL_METHYLATION_FILE, sep="\t", index=False)
```

```python
all_samples = pd.Series(list(filtered_supplemental_normal_methylation.drop(columns=['key']).columns) + list(sorted_filtered_samples_methylation.drop(columns=['id']).columns))
```

```python
all_cpgs = pd.Series(
    list(
        set(filtered_supplemental_normal_methylation['key']) | set(sorted_filtered_samples_methylation['id'])
        )
    )
```

```python
all_cpgs
```

```python
all_samples.to_csv(ALL_SAMPLES_FILE, index=False)
```

```python
all_cpgs.to_csv(ALL_CPGS_FILE, index=False)
```
