```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20, 20), nrows=1, ncols=2)

# Group by project_id and count the samples
grouped_samples_project = merged_meta_data.groupby("project_id").count()["sample"]
grouped_samples_project.name = ""

labels_project = [
    f"{label}\nN={count}"
    for label, count in zip(
        grouped_samples_project.index, grouped_samples_project.values
    )
]
grouped_samples_project.plot.pie(
    ax=ax[0], labels=labels_project, autopct="%1.1f%%", labeldistance=1.1, fontsize=18
)
ax[0].set_title("Samples by Project ID", fontsize=24)

# Group by sample_type and count the samples
grouped_samples_type = merged_meta_data.groupby("sample_type").count()["sample"]
grouped_samples_type.name = ""

labels_type = [
    f"{label}\nN={count}"
    for label, count in zip(grouped_samples_type.index, grouped_samples_type.values)
]

grouped_samples_type.plot.pie(ax=ax[1], labels=labels_type, autopct="%1.1f%%", fontsize=18)
ax[1].set_title("Samples by Sample Type", fontsize=24)

plt.show()
```
