```python
all_files = glob.glob("../data/raw_data/*/data_methylation_hm27_hm450_merged.txt")

meth_dfs = (pd.read_csv(f, sep="\t") for f in all_files)
meth_dfs = list(meth_dfs)
```

```python
all_files
```

```python
meth_dfs = list(meth_dfs) 
meth_data_df = meth_dfs[0][["ENTITY_STABLE_ID", "NAME", "DESCRIPTION", "TRANSCRIPT_ID"]]

for df in meth_dfs:
    meth_data_df = pd.merge(meth_data_df, df, on=["ENTITY_STABLE_ID", "NAME", "DESCRIPTION", "TRANSCRIPT_ID"], how="outer")

meth_data_df
```

```python
meth_data_df.to_csv('../data/methylation/meth_data.tsv', sep='\t', index=False)
```
